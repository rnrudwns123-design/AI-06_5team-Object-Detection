```python
import data_management as dm
import os
import shutil

# 기본 경로 설정
ROOT_DIR = dm.ROOT_DIR
DATA_DIR = dm.DATA_DIR
IMAGE_DIR = dm.IMAGE_DIR
ANNOT_DIR = dm.ANNOT_DIR

full_dict_path = dm.full_dict_path
err_txt_path = dm.err_txt_path
fixed_dict_path = dm.fixed_dict_path
no_dict_path = dm.no_dict_path

PARTIAL_DICT = dm.load_partial_dict()
NO_DICT = dm.load_no_dict()
```

```python
# 고칠 err 이미지를 to_roboflow 폴더로 복사
def copy_to_project_folder(data):
    os.makedirs(os.path.join(ROOT_DIR, "to_roboflow"))

    if type(data) == dict:
        data = list(data.keys())

    for path in data:
        from_path = os.path.join(IMAGE_DIR, path)
        to_path = os.path.join(ROOT_DIR, "to_roboflow", path)
        shutil.copyfile(from_path, to_path)
```

```python
copy_to_project_folder(NO_DICT)
```
