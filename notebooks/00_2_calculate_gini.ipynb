```python
"""
지니 계수(불평등도)를 계산하는 함수입니다.
FULL_DICT는 고정값이므로 계산할 수 없습니다. (0.493)
FIXED_DICT와 FINAL_DICT만 계산합니다.
"""

import data_management as dm

FIXED_DICT = dm.load_fixed_dict()

dm.calculate_gini(FIXED_DICT)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 클래스 균형 파악

pill_num_count_dict = dict()

for annot_list in FINAL_DICT.values():
    for dic in annot_list:
        pill_num = int(dic["label"])
        try:
            pill_num_count_dict[int(pill_num)] += 1
        except:
            pill_num_count_dict[int(pill_num)] = 1


pill_df = pd.DataFrame(columns=["pill_id", "count", "percentage"])

for key, value in pill_num_count_dict.items():
    pill_df.loc[len(pill_df)] = key, value, value/sum([len(a) for a in FINAL_DICT.values()]) * 100

pill_df = pill_df.sort_values("count")

# sns.barplot(pill_df, x="pill_id", y="percentage", hue="percentage")
# plt.title("전체 대비 클래스 비율")
# plt.show()
```
